## *Bunları doğrudan birleştiremeyiz çünkü bir tanesinde 100 bin veri varken diğerinde bin tane veri var. Bu yüzden merge metodunu kullanırız.*

```python
data = pd.merge(data,movies,on="item_id") # buradaki on neye göre birleştirmek istiyorsun sorusunu soruyor
# biz item_id dedik cünkü iki tabloda ortak olan sadece ikisi var
```

```python
data.head()
```

```python
table = data.pivot_table(index = "user_id",columns="title",values="rating")
table.head()
```

```python
starwars = table["Star Wars (1977)"]
starwars.head()
```

```python
# Starwars'a oylama yapanlar başka hangi filmlere ne kadar oylama yapmış bunlarla ilgili bir korelasyon oluşturucaz
correlation = table.corrwith(starwars)
```

```python
correlation
```

```python
corr = pd.DataFrame(correlation,columns=["Correlation"])
corr.dropna(inplace=True)
corr.head()
```

```python
data.drop((["timestamp"]),axis=1)
```

```python
# buradaki her filmin ortalama mean değerini bulalım tittle a göre gruplasın aynı zamanda ratingi göstersin
ratings = pd.DataFrame(data.groupby("title")["rating"].mean())
ratings.sort_values("rating",ascending=False).head()
#sort_values fonksiyonu, bir DataFrame veya Series'i belirli bir sütuna veya indekse göre sıralamak için kullanılır. Bu fonksiyon, verileri artan veya azalan düzende sıralamanıza olanak tanır.
```

```python
ratings["oysayisi"] = pd.DataFrame(data.groupby("title")["rating"].count())
ratings.head()
```

```python
# oy sayısı küçük olanları temizleyelim ancak bunu yapmadan önce bir sıralayalım
ratings.sort_values("oysayisi",ascending=False).head()
```

```python
corr = corr.join(ratings["oysayisi"])
corr.head()

# korelasyona oy sayısını ekledik
```

```python
corr[corr["oysayisi"]>100].sort_values("Correlation",ascending=False).head()
```
